# IFFT

```python
for slice in range(218, 300):
    input_dir = "DATASET_dicom_fourier_spectrum"
    output_dir = "DATASET_dicom_ifft"
    os.makedirs(output_dir, exist_ok=True)

    # Carregar o espectro de Fourier salvo (completo, incluindo fase)
    fshift = np.load(os.path.join(input_dir, f"fourier_slice_{slice}.npy"))
    

    # Aplicar ifftshift e transformada inversa
    f_ishift = np.fft.ifftshift(fshift)
    img_back = np.fft.ifft2(f_ishift)

    # Extrair a parte real
    img_back = np.abs(img_back).astype(np.uint8)

    # # Salvar imagem reconstruída
    reconstructed_path = os.path.join(output_dir, f"reconstructed_slice_{slice}.png")
    cv2.imwrite(reconstructed_path, img_back)
```
